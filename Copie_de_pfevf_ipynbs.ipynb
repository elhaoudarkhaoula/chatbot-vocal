```python
import os
import pandas as pd
import json
import re
import xml.etree.ElementTree as ET

root_folder = "/content/drive/MyDrive"

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[^\w\s]", "", text)
    return text.strip()

for root, dirs, files in os.walk(root_folder):
    for file in files:

        path = os.path.join(root, file)
        name, ext = os.path.splitext(file)

        try:

            # CSV
            if ext == ".csv":
                df = pd.read_csv(path, encoding="latin1", on_bad_lines="skip")
                df = df.drop_duplicates()
                df = df.fillna("")
                clean_path = os.path.join(root, name + "_clean.csv")
                df.to_csv(clean_path, index=False)
                print("CSV nettoyé :", file)

            # JSON
            elif ext == ".json":
                with open(path, "r", encoding="utf-8", errors="ignore") as f:
                    data = json.load(f)

                clean_path = os.path.join(root, name + "_clean.json")
                with open(clean_path, "w", encoding="utf-8") as f:
                    json.dump(data, f, indent=2)

                print("JSON nettoyé :", file)

            # XML
            elif ext == ".xml":
                tree = ET.parse(path)
                root_xml = tree.getroot()

                clean_path = os.path.join(root, name + "_clean.xml")
                tree.write(clean_path)

                print("XML nettoyé :", file)

            # TXT ou autres fichiers texte
            else:
                with open(path, "r", encoding="utf-8", errors="ignore") as f:
                    content = f.read()

                cleaned = clean_text(content)

                clean_path = os.path.join(root, name + "_clean.txt")
                with open(clean_path, "w", encoding="utf-8") as f:
                    f.write(cleaned)

                print("texte nettoyé :", file)

        except Exception as e:
            print("impossible traiter :", file)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/datasetColab/final_dataset.csv")

print(len(dataset))
dataset.head()
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(dataset["question"])
```

```python
def chatbot(question):

    q_vec = vectorizer.transform([question])

    similarity = cosine_similarity(q_vec, X)

    best_index = similarity.argmax()

    best_score = similarity[0][best_index]

    if best_score < 0.2:
        return "I could not find a good answer."

    return dataset.iloc[best_index]["answer"]
```

```python
import requests

def search_internet(question):

    try:
        # simplifier la question
        keywords = question.lower().replace("what is", "").replace("c'est quoi", "").replace("define", "")
        keywords = keywords.strip()

        # recherche wikipedia
        search_url = "https://en.wikipedia.org/w/api.php"

        params = {
            "action": "query",
            "list": "search",
            "srsearch": keywords,
            "format": "json"
        }

        response = requests.get(search_url, params=params)
        data = response.json()

        title = data["query"]["search"][0]["title"]

        # récupérer résumé
        summary_url = "https://en.wikipedia.org/api/rest_v1/page/summary/" + title

        response = requests.get(summary_url)
        summary = response.json()

        return summary["extract"]

    except:
        return "Désolé, je n'ai pas trouvé de réponse en ligne."
```

```python
import wikipedia

def search_wikipedia(question):

    search_url = "https://en.wikipedia.org/w/api.php"

    params = {
        "action": "query",
        "list": "search",
        "srsearch": question,
        "format": "json"
    }

    r = requests.get(search_url, params=params)
    data = r.json()

    if len(data["query"]["search"]) == 0:
        return None

    page_title = data["query"]["search"][0]["title"]

    summary_url = "https://en.wikipedia.org/api/rest_v1/page/summary/" + page_title

    r2 = requests.get(summary_url)

    if r2.status_code == 200:
        return r2.json().get("extract")

    return None
```

```python
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode

def record_audio():
  display(Javascript("""
    async function record() {
      const stream = await navigator.mediaDevices.getUserMedia({audio:true});
      const recorder = new MediaRecorder(stream);
      let chunks = [];

      recorder.ondataavailable = e => chunks.push(e.data);
      recorder.start();

      await new Promise(resolve => setTimeout(resolve, 5000));

      recorder.stop();

      return await new Promise(resolve => {
        recorder.onstop = async () => {
          const blob = new Blob(chunks);
          const buffer = await blob.arrayBuffer();
          const base64 = btoa(
            new Uint8Array(buffer)
              .reduce((data, byte) => data + String.fromCharCode(byte), '')
          );
          resolve(base64);
        };
      });
    }
  """))

  audio = output.eval_js("record()")
  audio_bytes = b64decode(audio)

  with open("audio.webm", "wb") as f:
      f.write(audio_bytes)

  print("Audio enregistré : audio.webm")

record_audio()
```

```python
# -----------------------
# INSTALLATION
# -----------------------
!pip install openai-whisper pandas langdetect deep-translator gTTS requests
!apt-get -y install ffmpeg


# -----------------------
# IMPORTS
# -----------------------
import whisper
import pandas as pd
import requests
from langdetect import detect
from deep_translator import GoogleTranslator
from gtts import gTTS
from IPython.display import Audio, display, Javascript
from google.colab import output
from base64 import b64decode
import subprocess
import re


# -----------------------
# GROQ API KEY
# -----------------------
GROQ_API_KEY = "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"


# -----------------------
# DATASET
# -----------------------
PATH_CSV="/content/drive/MyDrive/datasetColab/final_dataset.csv"

data = pd.read_csv(PATH_CSV).dropna(subset=["question","answer"])

questions = data["question"].str.lower().str.strip().tolist()
answers = data["answer"].tolist()


# -----------------------
# WHISPER MODEL
# -----------------------
model = whisper.load_model("small")


# -----------------------
# GROQ API
# -----------------------
def groq_api(query):

    url = "https://api.groq.com/openai/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {GROQ_API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "llama-3.1-8b-instant",
        "messages": [
            {
                "role": "system",
                "content": "You are a scientific assistant. Give clear and short definitions."
            },
            {
                "role": "user",
                "content": query
            }
        ],
        "temperature": 0.2
    }

    r = requests.post(url, headers=headers, json=payload)

    if r.status_code == 200:
        result = r.json()
        return result["choices"][0]["message"]["content"]

    return "I could not find information."


# -----------------------
# RECORD AUDIO
# -----------------------
def record_audio(filename="audio.webm", duration=5):

    print("🎤 Parlez maintenant (5 secondes)...")

    display(Javascript(f"""
    async function recordAudio(){{
        const stream = await navigator.mediaDevices.getUserMedia({{audio:true}});
        const recorder = new MediaRecorder(stream);
        const chunks=[];

        recorder.ondataavailable=e=>chunks.push(e.data);
        recorder.start();

        await new Promise(resolve=>setTimeout(resolve,{duration}000));

        recorder.stop();

        return new Promise(resolve=>{{
            recorder.onstop=async()=>{{
                const blob=new Blob(chunks);
                const reader=new FileReader();
                reader.readAsDataURL(blob);
                reader.onloadend=()=>resolve(reader.result.split(',')[1]);
            }}
        }})
    }}
    """))

    audio = output.eval_js("recordAudio()")

    with open(filename,"wb") as f:
        f.write(b64decode(audio))

    return filename


# -----------------------
# RECORD AUDIO
# -----------------------
audio_file = record_audio()


# -----------------------
# CONVERT AUDIO
# -----------------------
subprocess.run([
"ffmpeg","-y",
"-i",audio_file,
"-ac","1",
"-ar","16000",
"audio.wav"
], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)


# -----------------------
# TRANSCRIPTION
# -----------------------
result = model.transcribe("audio.wav")

user_text = result["text"].strip()

print("📝 Question :", user_text)


# -----------------------
# PROCESS QUESTION
# -----------------------
if not user_text:

    final_text = "Je n'ai pas compris la question."

else:

    # language detection
    try:
        user_lang = detect(user_text)
    except:
        user_lang = "fr"

    if user_lang.startswith("fr"):
        user_lang = "fr"
    elif user_lang.startswith("ar"):
        user_lang = "ar"
    elif user_lang.startswith("en"):
        user_lang = "en"
    else:
        user_lang = "fr"


    # translate to english
    try:
        query_en = GoogleTranslator(source='auto', target='en').translate(user_text)
        query_en = query_en.lower().strip()
    except:
        query_en = user_text.lower().strip()


    # -----------------------
    # DATASET EXACT SEARCH
    # -----------------------
    if query_en in questions:

        print("📚 Answer from dataset")

        index = questions.index(query_en)
        response_en = answers[index]

    else:

        print("🤖 Answer from Groq AI")

        response_en = groq_api(query_en)


    # -----------------------
    # TRANSLATE RESPONSE
    # -----------------------
    try:
        final_text = GoogleTranslator(source='en', target=user_lang).translate(response_en)
    except:
        final_text = response_en


# -----------------------
# PRINT RESPONSE
# -----------------------
print("🤖 Réponse :", final_text)


# -----------------------
# AUDIO RESPONSE
# -----------------------

# nettoyer le texte
clean_text = re.sub(r"[*#\-]", "", final_text)

tts = gTTS(text=clean_text, lang=user_lang)
tts.save("response.mp3")

display(Audio("response.mp3", autoplay=True))
```
